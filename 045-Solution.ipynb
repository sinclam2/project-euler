# Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:

## Triangle

### $T_{n}=n\frac{n+1}{2}$

#### 1, 3, 6, 10, 15, ...

## Pentagonal

### $P_{n}=n\frac{3n−1}{2}$

#### 1, 5, 12, 22, 35, ...

## Hexagonal

### $H_{n}=n(2n−1)$

#### 1, 6, 15, 28, 45, ...


# It can be verified that $T_{285} = P_{165} = H_{143} = 40755$

# Find the next triangle number that is also pentagonal and hexagonal.

```python
def triangle(n):
    return n*(n+1)/2.0
```

```python
def pentagonal(n):
    return n*(3*n-1)/2.0
```

```python
def hexagonal(n):
    return n*(2*n-1)
```

- Finding the first 100,000 of each type

```python
list_triangle_100000 = [triangle(n) for n in range(1, 100001)]
list_pentagonal_100000 = [pentagonal(n) for n in range(1, 100001)]
list_hexagonal_100000 = [hexagonal(n) for n in range(1, 100001)]
```

```python
(set(list_triangle_100000).intersection(set(list_pentagonal_100000))).intersection(set(list_hexagonal_100000))
```

# Answer: 1,533,776,805
