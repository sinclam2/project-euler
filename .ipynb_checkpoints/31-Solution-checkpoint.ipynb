# How many ways can we make 200p?

### The max number of each coin used in a combination is:

- 1p: 200
- 2p: 100
- 5p: 40
- 10p: 20
- 20p: 10
- 50p: 4
- 100p: 2
- 200p: 1

```python
list_combinations = []

for x_200 in [0,1]:
    total_so_far = 200*x_200
    if total_so_far==200:
        list_combinations.append([x_200])
        continue
    for x_100 in range((200-total_so_far)//100+1):
        total_so_far = 200*x_200+100*x_100
        if total_so_far==200:
            list_combinations.append([x_200, x_100])
            continue
        for x_50 in range((200-total_so_far)//50+1):
            total_so_far = 200*x_200+100*x_100+50*x_50
            if total_so_far==200:
                list_combinations.append([x_200, x_100,x_50])
                continue
            for x_20 in range((200-total_so_far)//20+1):
                total_so_far = 200*x_200+100*x_100+50*x_50+20*x_20
                if total_so_far==200:
                    list_combinations.append([x_200, x_100,x_50, x_20])
                    continue
                for x_10 in range((200-total_so_far)//10+1):
                    total_so_far = 200*x_200+100*x_100+50*x_50+20*x_20+10*x_10
                    if total_so_far==200:
                        list_combinations.append([x_200, x_100,x_50, x_20, x_10])
                        continue
                    for x_5 in range((200-total_so_far)//5+1):
                        total_so_far = 200*x_200+100*x_100+50*x_50+20*x_20+10*x_10+5*x_5
                        if total_so_far==200:
                            list_combinations.append([x_200, x_100,x_50, x_20, x_10, x_5])
                            continue
                        for x_2 in range((200-total_so_far)//2+1):
                            total_so_far = 200*x_200+100*x_100+50*x_50+20*x_20+10*x_10+5*x_5+2*x_2
                            if total_so_far==200:
                                list_combinations.append([x_200, x_100,x_50, x_20, x_10, x_5, x_2])
                                continue
                            for x_1 in range((200-total_so_far)+1):
                                total_so_far = 200*x_200+100*x_100+50*x_50+20*x_20+10*x_10+5*x_5+2*x_2+x_1
                                if total_so_far==200:
                                    list_combinations.append([x_200, x_100,x_50, x_20, x_10, x_5, x_2, x_1])
```
