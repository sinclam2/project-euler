# If p is the perimeter of a right angle triangle with integral length sides, {a,b,c}, there are exactly three solutions for p = 120.

# {20,48,52}, {24,45,51}, {30,40,50}

# For which value of p â‰¤ 1000, is the number of solutions maximised?

_____

### Let's start by thinking of a stick of length $p$

### How can we break the stick into 3 pieces (by snapping it twice) to make a right angle triangle?

### We know that for the three pieces to form a right angle triangle, no piece can be longer than the sum of the lengths of the other two pieces


### So we have the following conditions:

1. $a < b+c$
2. $b < a+c$
3. $c < a+b$

### But since we break a single stick, we know that $c = p - (a+b)$ so our conditions become:

1. $a < b+c \implies a < p/2$
2. $b < a+c \implies b < p/2$
3. $c < a+b \implies a+b > p/2$

## Let's visualize these conditions

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10

x = range(11)
y = [10-i for i in x]

plt.plot(x,y, 'k')
plt.plot([float(i)/2 for i in x],[float(i)/2 for i in y], 'k', alpha = 0.5)

plt.axvline(5, ls = 'dotted', color = 'r')
plt.axhline(5, ls = 'dotted', color = 'r')

plt.fill_between([float(i)/2 for i in x], 0, [float(i)/2 for i in y], color = 'k', alpha = 0.5)
plt.fill_between(x, y, 10, color = 'k')

plt.fill_between([float(i)/2 for i in x], 5, 10, color = 'k', alpha = 0.5)
plt.fill_between([5+float(i)/2 for i in x], 0, 5, color = 'k', alpha = 0.5)

my_ticks = ['0','','','','','p/2','','','','','p']
plt.xticks(x, my_ticks)
plt.yticks(x, my_ticks)
plt.xlim([0,10])
plt.ylim([0,10])
plt.xlabel('Length of a')
plt.ylabel('Length of b')

plt.title('Combinations of a and b values')
plt.grid()
plt.show()
```

## The combinations of $a$ and $b$ that lie inside the white area satisfy all the conditions

# Therefore, for a triangle with perimeter $p$, we loop through and count the possible $a$ and $b$ combinations

```python
max_count = 0
max_p = 0

for p in range(1,1001):
    count = 0
    max_ab = (p // 2) + 1
    for a in range(max_ab):
        min_b = max(a, max_ab - a)
        for b in range(min_b, max_ab):
            c = p - a - b
            #checking that it is indeed a right angle triangle
            if (c**2 == (a**2 + b**2)):
                count += 1
    if count > max_count:
        max_count = count
        max_p = p
```
