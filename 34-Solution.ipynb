# 145 is a curious number, as 1! + 4! + 5! = 1 + 24 + 120 = 145.

# Find the sum of all numbers which are equal to the sum of the factorial of their digits.

# Note: as 1! = 1 and 2! = 2 are not sums they are not included.

_____

# The maximum value for the factorial of a digit is 9! = 362880

# Therefore the maximum value for the sum of the factorials of the digits of an $n$ digit number is equal to $n\cdot9!$

# Let's plot the maximum value for a number of $n$ digits versus the maximum value of its factorial sum

## This will give us the upper limit on what to loop through

```python
from math import factorial
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
list_max = []
list_factorial = []

for n_digits in [1,2,3,4,5,6,7,8]:
    max_val = int('9'*n_digits)
    max_factorial = n_digits*factorial(9)
    
    list_max.append(max_val)
    list_factorial.append(max_factorial)
```

```python
x = range(1,9)
plt.plot(x, list_max, label = 'max')
plt.plot(x, list_factorial, label = 'max factorial')
plt.legend()
```

# It looks like after 6 digits, it's impossible for a number to be equal to the sum of the factorial of its digits

```python
list_results = []

#start the range at 3 since we know 1! and 2! are not allowed
for i in range(3,1000000):
    list_digits = [factorial(int(x)) for x in list(str(i))]
    if sum(list_digits)==i:
        list_results.append(i)
```

```python
list_results, sum(list_results)
```
