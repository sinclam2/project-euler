```python
df = pd.DataFrame(np.array([int(i) for i in x.split(' ')]).reshape(20,20))
```

## Getting max product for the columns and rows

```python
list_results = []
for col in df.columns:
    array = df[col].values
    for i in range(len(array)-4):
        temp_array = array[i:i+4]
        temp_max = 1
        for y in temp_array:
            temp_max *= y
        list_results.append(temp_max)
for row in df.index:
    array = df.loc[row].values
    for i in range(len(array)-4):
        temp_array = array[i:i+4]
        temp_max = 1
        for y in temp_array:
            temp_max *= y
        list_results.append(temp_max)
```

## Now, doing the diagonals

```python
df_diagonal = pd.DataFrame()

for row in range(20)[::-1]:
    list_temp_rows = [x for x in range(row,20)]
    list_temp_cols = [x for x in range(20-row)]
    list_indices= list(zip(list_temp_rows,list_temp_cols))
    list_values = [df.loc[row, column] for row, column in list_indices]
    df_diagonal = pd.concat([df_diagonal, pd.Series(list_values)], axis = 1, ignore_index = True)
    #flipping it to do the other half
    list_values = [df.loc[column, row] for row, column in list_indices]
    df_diagonal = pd.concat([df_diagonal, pd.Series(list_values)], axis = 1, ignore_index = True)
```

## Repeating the same procedure for columns from above

```python
for col in df_diagonal.columns:
    array = df_diagonal[col].values
    for i in range(len(array)-4):
        temp_array = array[i:i+4]
        temp_max = 1
        for y in temp_array:
            temp_max *= y
        list_results.append(temp_max)
```

## Now, using the diagonals in the other direction

```python
df2 = pd.DataFrame(columns = range(20))

for row in df.index:
    df2.loc[row] = list(df.loc[row])[::-1]
```

```python
df_diagonal2 = pd.DataFrame()

for row in range(20)[::-1]:
    list_temp_rows = [x for x in range(row,20)]
    list_temp_cols = [x for x in range(20-row)]
    list_indices= list(zip(list_temp_rows,list_temp_cols))
    list_values = [df2.loc[row, column] for row, column in list_indices]
    df_diagonal2 = pd.concat([df_diagonal2, pd.Series(list_values)], axis = 1, ignore_index = True)
    #flipping it to do the other half
    list_values = [df2.loc[column, row] for row, column in list_indices]
    df_diagonal2 = pd.concat([df_diagonal2, pd.Series(list_values)], axis = 1, ignore_index = True)
```

```python
for col in df_diagonal2.columns:
    array = df_diagonal2[col].values
    for i in range(len(array)-4):
        temp_array = array[i:i+4]
        temp_max = 1
        for y in temp_array:
            temp_max *= y
        list_results.append(temp_max)
```
