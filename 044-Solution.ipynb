# Pentagonal numbers are generated by the formula, $P_{n}=n\frac{3n-1}{2}$. The first ten pentagonal numbers are:

# 1, 5, 12, 22, 35, 51, 70, 92, 117, 145, ...

# It can be seen that $P_{4}+ P_{7} = 22 + 70 = 92 = P_{8}$. However, their difference, 70 − 22 = 48, is not pentagonal.

# Find the pair of pentagonal numbers, $P_{j}$ and $P_{k}$, for which their sum and difference are pentagonal and $D = |P_{k} − P_{j}|$ is minimised; what is the value of D?

```python
import numpy as np
```

```python
def pentagonal(n):
    return n*(3*n-1)/2.0
```

```python
def pentagonal_inverse(Pn):
    n = (1/6)*(1+np.sqrt(24*Pn + 1))
    return n
```

```python
def pentagonal_checker(x):
    n = pentagonal_inverse(x)
    if n%1==0:
        return True
    else:
        return False
```

```python
k = 0
found = False
while not found:
    k += 1
    for j in range(1,k):
        pj = pentagonal(j)
        pk = pentagonal(k)
        sum_val = pj+pk
        if pentagonal_checker(sum_val):
            diff_val = pk - pj
            if pentagonal_checker(diff_val):
                print(j,k)
                found = True
```
