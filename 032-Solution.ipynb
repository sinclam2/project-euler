# We shall say that an $n$-digit number is pandigital if it makes use of all the digits 1 to $n$ exactly once; for example, the 5-digit number, 15234, is 1 through 5 pandigital.

# The product 7254 is unusual, as the identity, 39 Ã— 186 = 7254, containing multiplicand, multiplier, and product is 1 through 9 pandigital.

# Find the sum of all products whose multiplicand/multiplier/product identity can be written as a 1 through 9 pandigital.

# *HINT: Some products can be obtained in more than one way so be sure to only include it once in your sum.*

_____

# First, we need to widdle down the number of products to loop through

### Let $f(n)$ denote the number of digits for the number $n$

### Then, if we're thinking about the product $x\cdot y = z$, then for it to be possible for the criteria to be satisfied, we need $f(x) + f(y) + f(z) = f(123456789) = 9$

### Let's start by thinking about a single digit $x$

### Then, we think about the possible number of digits for $y$ such that $f(x) + f(y) + f(x\cdot y)=9$

### The smallest possible value of $x$ is 1, and the greatest possible value is 9

### If $f(y) = 2$, the minimum $y$ value is 10 and the maximum value is 99

### Therefore, the minimum value for $z$ is $1\cdot11 = 11$ which has $f(11)=2$

### The maximum value for $z$ is $9\cdot99 = 891$ which has $f(891)=3$

### Therefore, if $f(x)=1$ and $f(y)=2$, $f(x)+f(y)+f(z)$ has a maximum value of 6, which means it's impossible that any of these products will satisfy the condition

# We can plot these max and min $f(z)$ values to see the combinations

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
min_digits = []
max_digits = []

for n_digits in [1,2,3,4,5,6]:
    min_val = 1*(int('1'+'0'*(n_digits-1)))
    max_val = 9*(int('9'*n_digits))
    min_digits.append(len(str(min_val))+1+n_digits)
    max_digits.append(len(str(max_val))+1+n_digits)
```

```python
plt.plot(range(1,7),min_digits, label = 'min f(x)+f(y)+f(z)')
plt.plot(range(1,7),max_digits, label = 'max f(x)+f(y)+f(z)')
plt.axhline(9, ls = 'dotted', color = 'r')
plt.xlabel('f(y)')
plt.ylabel('f(x)+f(y)+f(z)')
plt.title('f(x)=1')
plt.legend()
```

# As we can see, the only possibility is $f(y)=4$

```python
list_products = []
```

```python
for x in range(1,10):
    for y in range(1000,10000):
        z = x*y
        combined = str(x)+str(y)+str(z)
        if (len(combined)==9)&(len([i for i in range(1,10) if str(i) in combined])==9):
            print('{} x {} = {}'.format(x,y,z))
            list_products.append(z)
```
